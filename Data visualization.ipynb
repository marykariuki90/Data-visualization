# loading the dataset using seaborn
we use the load dataset() function in seaborn to import the dataset where we use dataset called tips. Tips is a dataset collecteded on a restuarant

```python
data=sns.load_dataset("tips")
data.head()
```

# correlation heatmap

# DISTRIBUTION PLOTS
There are graphs that uses numerical data to make plots they include

distplot=univariate analysis(only one feature)

jointplot=bivariate analysis(comparison between two features)

paiplot=used where you have more than two features

# join-plot
a joint plot is used in bavariate analysis  and univariate analysis plotted together where it comprises of three plots.
the first plot which is the main plot shows how dependent variable (y) varies with the  independent variable (x).
The second plot placed horizontally at the top shows distribution of the independent variable (x). And the third plot placed vertically at the right shows the distribution of the dependent variable(y)


```python
#kind means that instead of points at the main plot they will be hexagon
sns.jointplot(x="total_bill",y="tip",data=data,kind="hex")
```

# kind= reg
reg for regression,which draws a regression line(line of the best fit) at the main plot and at marginal plot it will draw a probability density function

```python
sns.jointplot(x="total_bill",y="tip",data=data,kind="reg")
```

# pairplot
it is used when you have more than two features, a pairplot iswhere one variable in the same data row is matched with another variables values

```python
sns.pairplot(data)
```

```python
#the results will be based on sex column
sns.pairplot(data,hue="sex")
```

```python
#the results shows most  the people were none smoker
sns.pairplot(data,hue="smoker")
```

# distplot
it is used to check trhe distribution of the column feature since it is meant for univariate analysis

```python
sns.distplot(data["tip"])
```

```python
sns.distplot(data["total_bill"],kde=False)
```

```python
sns.distplot(data["size"],kde=False,bins=10)
```

# categorical plots
seaborn provides a way of drawing plots on data that is categorical such as object,category

# countplot
it shows the count of the observation in each categorical bin using bars

```python
#sex by default is always taken as x-axis
sns.countplot("sex",data=data)
```

```python
#day is taken as default x-axis therefore getting a vertical plot
sns.countplot("day",data=data)
```

```python
#you can change the default x-axis to y-axis and get a horizontal plot
sns.countplot(y="smoker",data=data)
```

# barplot
in barplot you must specify the x-axis and y-axis unlike in countplot where you only specify one axis since the rest will be the count

```python
sns.barplot(y="total_bill",x="sex",data=data)
```

# boxplot
it is a graph that represent information from five number summary starting at the lowest that is 0%,25%,50%,75% and the points above the boxplot indicates the outliers

```python
sns.boxplot("sex","total_bill",data=data)
```

```python
#if you do not include the palette it will use its default colors
sns.boxplot(x="day",y="total_bill",data=data,palette="rainbow")
```

```python
#boxplot for all the numeric columns since we have include the orient
sns.boxplot(data=data,orient='v')
```

```python
sns.boxplot(x="day",y="total_bill",hue="smoker",data=data)
```

# violinplot
it is a graph that help us to see both the distribution of the datain terms of kernel density estimation and boxplot,that is inside the violin plot there is a boxplot

```python
sns.violinplot(x="sex",y="total_bill",data=data,palette="rainbow")
```

```python
sns.violinplot(x="day",y="total_bill",data=data,palette="rainbow")
```
